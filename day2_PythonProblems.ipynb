```python
### Number Comparision Tool
###step 1: Get two numbers from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
###step 2: Compare the numbers and display the result
print("------ Comparison Result ----")
if num1 > num2:
    print(f"{num1} is greater than {num2}")
elif num1 < num2:
    print(f"{num1} is less than {num2}")
else:
    print(f"{num1} is equal to {num2}")

##step 3: if any one number is zero
if num1 == 0 or num2 == 0:
    print("One of the numbers is zero.")
else:
    print("Neither number is zero.")
### A simple program to compare two numbers and check for zero.
```

```python
### Countdown Timer
import time
###step 1: Get the countdown time from the user
countdown_time = int(input("Enter countdown time in seconds: "))
###step 2: Start the countdown
print("------ Countdown Timer ----")
while countdown_time > 0:
    print(f"Time left: {countdown_time} seconds")
    time.sleep(1)
    countdown_time -= 1
print("Happy New Year!")
```

```python
### Basic Math quiz game
import random

### Define the math question function
def genarate_question():
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    operation = random.choice(['+', '-', '*', '/'])
    
    if operation == '+':
        answer = num1 + num2
    elif operation == '-':
        answer = num1 - num2
    elif operation == '*':
        answer = num1 * num2
    else:
        answer = round(num1 / num2, 2)  # rounding for division
    
    return f"What is {num1} {operation} {num2}?", answer

###step 1: Main quiz game
def math_quiz():
    score = 0
    total_questions = 5
    print("------ Math Quiz Game ----")
    print("Answer the following questions:")
    for _ in range(total_questions):
        question, correct_answer = genarate_question()
        user_answer = float(input(question + " "))
        
        if user_answer == correct_answer:
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer is {correct_answer}")

    print
    print(f"Your final score is {score} out of {total_questions}")

    if score == total_questions:
        print("Excellent! You got all questions right!")
    else:
        print("Good effort! Keep practicing to improve your score.")

### step 2: run the quiz game
math_quiz()
```
